```python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}

    # установка оценки лектору конкретным студентом
    def rate_hw(self, lecturer, course, grade):
        if isinstance(lecturer, Lecturer) and course in self.courses_in_progress and course in lecturer.courses_attached:
            if course in lecturer.grades:
                lecturer.grades[course] += [grade]
            else:
                lecturer.grades[course] = [grade]
        else:
            return 'Ошибка'

    # перегруженный метод print
    def __str__(self):
        return "Имя: " + self.name + "\n" + "Фамилия: " + self.surname + "\n" + "Средняя оценка за домашние задания: " + str(avg_dict(self.grades)) + "\n" + "Курсы в процессе изучения: " + ", ".join( self.courses_in_progress) + "\n" + "Завершённые курсы: " + ", ".join(self.finished_courses)


    # перегрузка оператора <
    def __lt__(self, student):
        avg_student_1 = avg_dict(self.grades)
        avg_student_2 = avg_dict(student.grades)
        if avg_student_1 < avg_student_2:
            return True
        else:
            return False


class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []

class Lecturer(Mentor):

    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []
        # оценки конкретному лектору
        self.grades = {}

    # перегруженный метод print
    def __str__(self):
        return "Имя: " + self.name + "\n" + "Фамилия: " + self.surname + "\n" + "Средняя оценка за лекцию: " + str(avg_dict(self.grades))

    # перегрузка оператора <
    def __lt__(self, lecturer):
        avg_lecturer_1 = avg_dict(self.grades)
        avg_lecturer_2 = avg_dict(lecturer.grades)
        if avg_lecturer_1 < avg_lecturer_2:
            return True
        else:
            return False

class Reviewer(Mentor):

    def rate_hw(self, student, course, grade):
        if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            if course in student.grades:
                student.grades[course] += [grade]
            else:
                student.grades[course] = [grade]
        else:
            return 'Ошибка'

    # перегруженный метод print
    def __str__(self):
        return "Имя: " + self.name + "\n" + "Фамилия: " + self.surname

# среднее арифметическое по словарю
def avg_dict(dict1):
    summ = 0
    count = 0
    for elem in dict1.values():
        for elem2 in elem:
            summ = summ + elem2
            count = count + 1

    if count!=0:
        return summ/count
    else:
        return 0

# среднее арифсметическое по всем оценкам студентов по определённому курсу
def avg_students(students,course):
    summ = 0
    count = 0
    # проход  по всем студентам
    for student in students:
        # проход по всем оценкам конкретного студента
        for cour, grade in student.grades.items():
            # если курс, за которую студент поулчил оценку равен курсу, который нам нужен
            if cour == course:
                for grade1 in grade:
                    # добавляем оценку в общую сумму
                    summ = summ + grade1
                    count = count + 1

    if count != 0:
        return summ / count
    else:
        return 0

# среднее арифсметическое по всем оценкам лекторов по определённому курсу
def avg_lecturers(lecturers, course):
    summ = 0
    count = 0
    # проход  по всем лекторам
    for lecturer in lecturers:
        # проход по всем оценкам конкретного студента
        for cour, grade in lecturer.grades.items():
            # если курс, за которую лектор поулчил оценку равен курсу, который нам нужен
            if cour == course:
                for grade1 in grade:
                    # добавляем оценку в общую сумму
                    summ = summ + grade1
                    count = count + 1

    if count != 0:
        return summ / count
    else:
        return 0


# Два студента
best_student1 = Student('Ruoy', 'Eman', 'man')
best_student1.courses_in_progress += ['Python']

best_student2 = Student('Igor', 'Ivanov', 'man')
best_student2.courses_in_progress += ['Python']

# Два лектора
cool_lec1 = Lecturer('Roman', 'Buddy')
cool_lec1.courses_attached += ['Python']

cool_lec2 = Lecturer('Andrew', 'Petrov')
cool_lec2.courses_attached += ['Python']

# Два ревьвера
cool_rev1 = Reviewer('Roman', 'Buddy')
cool_rev1.courses_attached += ['Python']

cool_rev2 = Reviewer('Andrew', 'Petrov')
cool_rev2.courses_attached += ['Python']

# всё для ревьюверов
cool_rev1.rate_hw(best_student1, 'Python', 10)
print(cool_rev1)

cool_rev1.rate_hw(best_student2, 'Python', 2)

# всё для студентов
best_student1.rate_hw(cool_lec1, 'Python', 10)
best_student1.rate_hw(cool_lec1, 'Python', 6)
print(best_student1)
print(best_student1 < best_student2)

# всё для лектора
print(cool_lec1 < cool_lec2)
print(cool_lec1)

#Среднее арифметическое всех студентов по определённому курсу
result1 = avg_students([best_student1, best_student2],"Python")
print("Среднее арифметическое всех студентов: " + str(result1))

#Среднее арифметическое всех студентов по определённому курсу
result2 = avg_lecturers([cool_lec1, cool_lec2],"Python")
print("Среднее арифметическое всех лекторов: " + str(result2))
```
